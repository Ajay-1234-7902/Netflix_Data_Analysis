```python
loanX = loan.drop(['Customer_Age_Bucket'], axis=1)
```

```python
loanX['Loan Taken Year'] = pd.to_datetime(loan['Loan Taken Date']).dt.year
loanX['Loan Taken Month'] = pd.to_datetime(loan['Loan Taken Date']).dt.month
loanX['Loan Taken Day'] = pd.to_datetime(loan['Loan Taken Date']).dt.day
```

```python
loanX.drop('Loan Taken Date', inplace=True, axis=1)
```

```python
loanX.drop('Customer ID', inplace=True, axis=1)
```

```python
loanX.drop('Fake_Neighbour_Names', inplace=True, axis=1)
```

```python
categoric_cols.columns
```

```python
customer_age = pd.get_dummies(categoric_cols['Customer_Age_Bucket'], prefix="Age_Bucket")
```

```python
loan_taken_for = pd.get_dummies(categoric_cols['Loan Taken For'], prefix="loan_taken_for")
```

```python
loan_type = pd.get_dummies(categoric_cols['Loan Type'], prefix="loan_type")
```

```python
loan_rate_type = pd.get_dummies(categoric_cols['Loan Rate Type'], prefix="loan_rate_type")
```

```python
pre_pay_charge = pd.get_dummies(categoric_cols['Pre Payement Charges'], prefix="pre_payment_charges")
```

```python
city = pd.get_dummies(categoric_cols['City'], prefix="city")
```

```python
job_type = pd.get_dummies(categoric_cols['Job type'], prefix="job_type")
```

```python
has_other_loans = pd.get_dummies(categoric_cols['Has Other Loans'], prefix="has_other_loans")
```

```python
emi_pay_mode = pd.get_dummies(categoric_cols['EMI Payement Mode'], prefix="emi_payment_mode")
```

```python
X = loanX.drop(['Loan Taken For', 'Loan Type', 'Loan Rate Type',
       'Pre Payement Charges', 'City', 'Job type', 'Has Other Loans',
       'EMI Payement Mode'], axis=1)
```

```python
X = pd.concat([X, customer_age, loan_taken_for, loan_type, loan_rate_type, pre_pay_charge, city, job_type, has_other_loans, emi_pay_mode], axis=1)
```

```python
X.drop(['First Defaulted On', 'last Defaulted On', 'No. Of 1 Day Delay', 'No Of 2 Days Delay'], inplace=True, axis=1)
```

```python
y = X['Percentage of Yield']
```

```python
X.drop('Percentage of Yield', inplace=True, axis=1)
```

```python
X.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, pred)
```

```python
r2
```
